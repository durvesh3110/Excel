1. **What is the difference between 'Paste' and 'Paste Special' in Excel?** **Briefly explain with examples.**

**Ans:**

1.   Paste
  - Copies everything exactly as it is (values, formulas, formatting, comments).
  - Shortcut: Ctrl + V
  - Example:

    If cell A1 contains =B1+C1 with bold formatting, normal Paste will copy:

      - The formula
      - The formatting
      - The result

2.   Paste Special
  - Allows you to choose what part of the copied data to paste.
  - Common options:
    - Values only
    - Formulas only
    - Formatting only
    - Transpose (rows → columns)
  - Example:
    - Copy a formula.
    - Use Paste Special → Values to paste only the calculated result (not the formula).
  - Useful when you want to remove formulas but keep results.

    

2. **Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes'** **in Excel.**

**Ans:**



1.   Freze Panes
  - Function
    - Locks specific rows and/or columns so they remain visible while scrolling through the worksheet.
  - Uses
    - Keeps header rows visible in large datasets.
    - Keeps the first column visible when working with wide tables.
    - Helps maintain reference labels while navigating large amounts of data.
    - Useful in reports, financial sheets, and data analysis tasks.

2.   Split panes

  - Function:
    - Divides the worksheet into multiple independently scrollable sections.
  - Uses:
    - Compare different parts of the same worksheet at the same time.
    - View distant rows or columns simultaneously.
    - Analyze large datasets without losing sight of other sections.
    - Useful for reviewing or auditing data.

3. **Explain the difference between inserting a new row and inserting a new column in Excel. Can you insert multiple rows or columns at once?**

**Ans:**

  - Inserting a New Row
    - What it does:
      Adds a horizontal row to the worksheet.

    - Existing rows shift downward.

    - Example:
      - If you insert a row above Row 5, the old Row 5 becomes Row 6.
    - Used when:
      - Adding a new record (e.g., new employee, new product entry).

  - Inserting a New Column
    - What it does: Adds a vertical column to the worksheet.
    - Existing columns shift to the right.
    - Example:
      - If you insert a column before Column B, the old Column B becomes Column C.

    - Used when:
      - Adding a new field (e.g., Salary column in an employee table).

  - Yes, you can insert multiple rows and columns at once.

5. **Discuss the purpose of 'XLOOKUP' and how it differs from the traditional 'VLOOKUP' function.**

**Ans :**

  - VLOOKUP (Traditional)
    - Searches vertically.
    - Can only search left → right.
    - Requires column index number.
    - Example:

      =VLOOKUP(A2, A1:C10, 2, FALSE)

  - XLOOKUP (Modern & Better)
    - Works both vertical & horizontal.
    - Can search in any direction.
    - No column index needed.
    - Handles errors better.
    - Example:

      =XLOOKUP(A2, A1:A10, B1:B10)
